```python
import wfdb
import os
from matplotlib import pyplot as plt
```

```python
#download MIT-BIH databaset
wfdb.dl_database('mitdb', (os.getcwd() + "/mitdb/"))
```

```python
# read a sample record from the mit-bih dataset with annotations

record = wfdb.rdrecord('/content/mitdb/100', sampto=2000)
annotation = wfdb.rdann('/content/mitdb/100', 'atr', sampto=2000)
```

```python
# visualise the sample read from the dataset

wfdb.plot_wfdb(record=record, annotation=annotation, plot_sym=True, time_units='seconds', title='MIT-BIH Record 100', figsize=(15,4), ecg_grids='all')
```
